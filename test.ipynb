```python
import torch
import torch.nn as nn

# Example sentence
tokens = ["I", "love", "cats"]

# Dummy vocab and token to index
vocab = {"I": 0, "love": 1, "cats": 2}
token_ids = torch.tensor([vocab[token] for token in tokens])  # [0, 1, 2]

# Parameters
vocab_size = len(vocab)
embedding_dim = 8  # small for demo
max_len = 10       # max sequence length

# Token embedding layer
token_embedding = nn.Embedding(vocab_size, embedding_dim)

# Positional embedding layer
positional_embedding = nn.Embedding(max_len, embedding_dim)

# Get token embeddings
token_embed = token_embedding(token_ids)  # shape: [seq_len, embedding_dim]

# Get position ids
position_ids = torch.arange(len(tokens))  # [0, 1, 2]
position_embed = positional_embedding(position_ids)

# Add token + positional embeddings
input_to_transformer = token_embed + position_embed

# Print shapes and values
print("Token Embeddings:\n", token_embed)
print("Positional Embeddings:\n", position_embed)
print("Final Input Embeddings:\n", input_to_transformer)
```
